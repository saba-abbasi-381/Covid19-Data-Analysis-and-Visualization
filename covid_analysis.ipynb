## <center>**Data CLeaning**</center>

### **Detect Missing values in each Column of Covid 19 Dataset**

```python
df.isnull().sum()
```

### **Percentage of all Missing Values of each Column**

```python
df.isnull().mean()*100
```

### **Funuction for Fill Missing Values in each Column**

```python
def impute_null_values(data):
    for column in data.columns:
        if column in data.select_dtypes(include=['int64','int32','float64','float32']).columns:
            data[column].fillna(data[column].median(),inplace=True)
        elif column in data.select_dtypes(include='object').columns:
            data[column].fillna(data[column].mode()[0],inplace=True)
        else:
            print("column belongs to irrelevent data type")
impute_null_values(df)
print("Null values Impute Successfully")
```

### **After filling Missing values in each Column**

```python
df.isnull().sum()
```

## <center>**Data Visualization of Covid 19 Dataset**</center>

### **Relation of Population with Tests**

```python
sns.lineplot(x=df['population'],y=df['Tests'],lw=3,color='g')
plt.grid()
title_dict={
    "family":"Algerian",
    "size":20,
    "color":"darkred",
    "weight":"bold"
}
label_dict={
    "family":"Arial",
    "size":16,
    "color":"black",
    "weight":"bold"
}
plt.title('Relation of Population with Tests',fontdict=title_dict)
plt.xlabel('Population',fontdict=label_dict)
plt.ylabel('Tests',fontdict=label_dict)
plt.show()
```

### **Insights**
- The graph shows a positive relationship between population size and number of tests conducted.
- Higher population regions perform more tests, including testing scales with population demand.
 

### **Relation of Cases with Tests**

```python
sns.lineplot(x=df["Tests"],y=df["Cases"],lw=3,color='g')
plt.grid()
title_dict={
    "family":"Algerian",
    "size":20,
    "color":"darkred",
    "weight":"bold"
}
label_dict={
    "family":"Arial",
    "size":16,
    "color":"black",
    "weight":"bold"
}
plt.title('Relation of test with Cases',fontdict=title_dict)
plt.xlabel('Tests',fontdict=label_dict)
plt.ylabel('Cases',fontdict=label_dict)
plt.show()
```

### **Insights**
- Increase in testing leads to higher reported cases, showing that case numbers depend strongly on testing levels.

### **Population based on Continent** 

```python
sns.barplot(x=df["continent"],y=df["population"],hue=df['continent'])
title_dict={
    "family":"Algerian",
    "size":20,
    "color":"darkred",
    "weight":"bold"
}
label_dict={
    "family":"Arial",
    "size":16,
    "color":"black",
    "weight":"bold"
}
plt.title("Population based on Continent",fontdict=title_dict)
plt.xlabel("Continent",fontdict=label_dict)
plt.ylabel("Population",fontdict=label_dict)
plt.xticks(rotation=45)
plt.show()
```

### **Percentage of Population based on Continent**

```python
x=df.groupby('continent')['population'].sum()
plt.pie(
    x.values,
    labels=x.index,
    autopct='%.2f%%'
)
plt.show()
```

* 🥇 **Asia** accounts for the largest share of the world population.
* 🌏 Other continents like Africa, Europe, and the Americas have smaller shares.
